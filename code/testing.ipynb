```python
from rdflib import Graph
import networkx as nx
import matplotlib.pyplot as plt
import os

from urllib.parse import urlparse

def shorten(uri):
    if "#" in uri:
        return uri.split("#")[-1]
    elif "/" in uri:
        return uri.rstrip("/").split("/")[-1]
    else:
        return uri


# Pfad zur externen TTL-Datei (hier als Platzhalter)
ttl_file_path = "generated_books_mini.ttl"

# Prüfen, ob die Datei vorhanden ist
if os.path.exists(ttl_file_path):
    # RDF-Graph erstellen und TTL-Datei laden
    g = Graph()
    g.parse(ttl_file_path, format="turtle")

    # NetworkX-DiGraph zur Visualisierung
    G = nx.DiGraph()

    # Tripel in NetworkX übernehmen
    for subj, pred, obj in g:
        G.add_edge(shorten(str(subj)), shorten(str(obj)), label=shorten(str(pred)))
        #print(subj, pred, obj)

    # Positionen für Knoten berechnen
    pos = nx.spring_layout(G, k=0.5, iterations=50)

    # Knoten und Kanten zeichnen
    plt.figure(figsize=(12, 8))
    nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=1500, font_size=10, font_weight="bold", arrows=True)

    # Labels für Kanten anzeigen
    edge_labels = nx.get_edge_attributes(G, 'label')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

    plt.title("RDF Graph")
    plt.axis("off")
    plt.show()

else:
    print("Keine TTL-Datei gefunden. Bitte lade eine Datei hoch mit dem Namen 'beispiel.ttl'.")


# Funktion zur Berechnung aller Pfade von einem Root-Knoten aus
def find_all_paths(G, start_node):
    # Hilfsfunktion zur Tiefensuche (DFS)
    def dfs(node, path):
        # Wenn der Knoten bereits im Pfad ist, überspringen (Vermeidung von Zyklen)
        if node in path:
            return
        
        # Füge den aktuellen Knoten zum Pfad hinzu
        path.append(node)
        
        # Wenn der Knoten keine Nachbarn mehr hat, gibt es keinen weiteren Pfad
        neighbors = list(G.neighbors(node))
        if not neighbors:
            paths.append(list(path))  # Speichere den aktuellen Pfad
        
        # Rekursiv für jeden Nachbarn den Pfad erweitern
        for neighbor in neighbors:
            dfs(neighbor, path)
        
        # Entferne den Knoten wieder aus dem Pfad, um den nächsten Pfad zu finden
        path.pop()

    # Liste, um alle Pfade zu speichern
    paths = []
    
    # Starte die Rekursion von 'start_node'
    dfs(start_node, [])
    
    return paths

# Zähle alle Pfade von 'A'
start_node = "Node158"
all_paths_158 = find_all_paths(G, start_node)
all_paths_124 = find_all_paths(G, "Node124")

for paths in all_paths_124:
    all_paths_158.append(paths)



# Ausgabe der gefundenen Pfade
print("Gefundene Pfade:")
for path in all_paths_124:
    print(path)
```

```python
# Degree Distribution (Formula in: source 37, page 7)
# nG...number of nodes in gaph
# nE...number of edges in graph

# Calculating nE
query = """
SELECT (COUNT(*) AS ?tripleCount)
WHERE {
  ?s ?p ?o .
}
"""

res = g.query(query)
for row in res:
    print("Number of edges in graph: " + str(row['tripleCount']))
    nE = int(row['tripleCount'])

# Calculating nG
query = """
SELECT (COUNT(DISTINCT ?node) AS ?nodeCount)
WHERE {
  {
    SELECT ?node WHERE {
      { ?node ?p ?o }       
      UNION
      { ?s ?p ?node }       
    }
  }
}
"""

res = g.query(query)
for row in res:
    print("Number of nodes in graph: " + str(row['nodeCount']))
    nG = int(row['nodeCount'])

# Calculating degree for every node in graph
query = """
SELECT ?node (COUNT(?any) AS ?degree)
WHERE {
  {
    { ?node ?p ?any }     # Outgoing edges
    UNION
    { ?any ?p ?node }     # Incoming edges
  }
}
GROUP BY ?node
"""

# sum_of_degress = 0
degrees = []

res = g.query(query)
for row in res:
    print("Node: " + str(row['node']) + " - Degree: " + str(row['degree']))
    #sum_of_degress += int(row['degree'])
    degrees.append(int(row['degree']))
   
print("Sum of Degrees: " + str(sum(degrees)))

if nG > 1:
    mean_degree = (2 * nE) / nG
    squared_diffs = [(deg_v - mean_degree) ** 2 for deg_v in degrees]
    degree_deviation = sum(squared_diffs) / (nG-1)
else:
    degree_deviation = 0 

print("Degree Deviation: " + str(degree_deviation))
```
